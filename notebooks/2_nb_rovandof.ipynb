```python
#1.- Cargue el archivo dolar_2019_sii.csv a un dataframe que se llame dolar_2019 usando pandas.
#¿Que argumento necesita pd.read_csv() para leer las columnas correctamente?
```

para leer los datos de tamaño de fila variable necesitamos o generar el header de antemano ,aqui utilizando el read csv dos veces para leer el header primero y asi que pandas sepa que siempre tiene que leer el mismo tamaño de fila, sino tirara un error, ademas se utiliza skipfooter para no leer los promedios, porque es algo que podemos generar con pandas, y seteamos el indice del dataframe en la columna "Día"

```python
import pandas as pd

dolar_2019 = pd.read_csv('../data/raw/dolar2019_sii.csv',index_col = 'Día',delimiter=";", skipfooter=1,skiprows=1, names=list(pd.read_csv('../data/raw/dolar2019_sii.csv',delimiter=";",header='infer',nrows=1)))
dolar_2019.fillna(0, inplace=True)

dolar_2019.head(31)
```

```python

#2.-Explore el archivo dolar_2020_sii.txt. ¿Qué estructura tiene, en que se diferencia
#y en qué es igual al dataframe anterior? ¿Qué método y argumento hay que usar para leer
#las columnas correctamente?. Carguelo en un dataframe llamado dolar_2020_txt.
```

```python
#el archivo es distinto en los separadores y en que no hay una fila de headers
#volvemos a usar el read csv, cambiando el delimiter, saltando el footer
#y reusamos el header del archivo anterior porque son la misma cantidad de datos , para que queden
#dos dataframes con columnas uniformes :D
dolar_2020_txt = pd.read_csv('../data/raw/dolar_2020_sii.txt',index_col = 0,delimiter="|", skipfooter=1, names=list(pd.read_csv('../data/raw/dolar2019_sii.csv',delimiter=";",header='infer',nrows=1)))
dolar_2020_txt.fillna(0, inplace=True)

dolar_2020_txt.head(31)
```

```python
#3.-Explore el archivo dolar_2020_sii_AA.txt. ¿Qué estructura tiene? 
#¿Será buena idea ocupar expresiones regulares para indicar el separador? (hint: sí).
#Carguelo en un dataframe llamado dolar_2020_regex y compárelo con el dataframe dolar_2020_txt.
#¿Son exactamente iguales en la información que traen?
```

```python
#la estructura se ve similar a la del archivo dolar_2020.txt pero los separadores son mas complejos
#asi que habra que setear el delimiter con un regex
dolar_2020_regex = pd.read_csv('../data/raw/dolar_2020_sii_AA.dat',index_col = 0,delimiter="[a-zA-Zá-ý]{3}", skipfooter=1, names=list(pd.read_csv('../data/raw/dolar2019_sii.csv',delimiter=";",header='infer',nrows=1)))
dolar_2020_regex.fillna(0, inplace=True)

dolar_2020_regex.head(31)
```

```python
#con respecto a si traen o no la misma informacion, con la funcion compare podemos ver que no hay diferencias
#entre ambos dataframes
dolar_2020_regex.compare(dolar_2020_txt)
```

```python
#4.-Haga las siguientes operaciones de wrangling y limpieza:

#a) Filtre sacando la última fila, arreglando y homogeneizando los encabezados para los tres dataframes

#R: ya se hizo con skiprow al momento de leerla,
#y reuse los encabezados del primer archivo para homogeneizar

```

```python
#b) Revise que los nombres de las columnas sean los mismos en cada dataframe.
print(dolar_2019.columns,dolar_2020_txt.columns,dolar_2020_regex.columns)
#como se puede ver aqui las columnas son las mismos en todos los dataframe.
```

```python
#c) Cree una columna en cada uno de los tres dataframes que se llame year y que tenga el año correspondiente
#, y que su type sea int64.
dolar_2020_regex['year']=2020
dolar_2020_txt['year']=2020
dolar_2019['year']=2019

print("----listado de columnas---")
print(dolar_2019.columns,dolar_2020_txt.columns,dolar_2020_regex.columns)
print("----imprimo tipo de columna---")
print(dolar_2019.year.dtype,dolar_2020_txt.year.dtype,dolar_2020_regex.year.dtype)
```

```python
#d) Convierta las comas a puntos en cada uno de los tres dataframes.

cols=[i for i in dolar_2019.columns if i not in ["year"]]
for col in cols:
    dolar_2019[col]=dolar_2019[col].str.replace(',','.')
    dolar_2020_txt[col]=dolar_2020_txt[col].str.replace(',','.')
    dolar_2020_regex[col]=dolar_2020_regex[col].str.replace(',','.')

dolar_2019.fillna(0, inplace=True)
dolar_2020_regex.fillna(0, inplace=True)
dolar_2020_txt.fillna(0, inplace=True)


print(dolar_2019.head(3)
,dolar_2020_regex.head(3)
,dolar_2020_txt.head(3))
```

```python
#e) Fuerce que todas las variables sean de tipo int o de tipo float.

cols=[i for i in dolar_2019.columns if i not in ["year"]]
for col in cols:
    dolar_2019[col]=pd.to_numeric(dolar_2019[col])
    dolar_2020_txt[col]=pd.to_numeric(dolar_2020_txt[col])
    dolar_2020_regex[col]=pd.to_numeric(dolar_2020_regex[col])


print("----imprimo tipo de columnas---")

print(dolar_2019.dtypes,dolar_2020_txt.dtypes,dolar_2020_regex.dtypes)
```
